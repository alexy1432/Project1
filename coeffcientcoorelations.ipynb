```python
# Dependencies and Setup
import os
import pandas as pd
import csv
import scipy.stats as stats
import matplotlib
from matplotlib import pyplot as plt
```

```python
#to read coeffecint of correlations between education and mortality
Mortality_df = pd.DataFrame(pd.read_csv("CleanedStateSib.csv"))
Mortality_df.head()
mortality2014 = Mortality_df.loc[Mortality_df['Year'] == 2014]
#df.loc[df['column_name'] == some_value]
mortality2014.head()
```

```python
Obesity_df = pd.DataFrame(pd.read_csv("CleanedObesitystateCorr.csv"))
Obesity_df.head()
```

```python
Obesity_df2014 = Obesity_df[['State','Adult Obesity Rate 2014']]
Obesity_df2014 = Obesity_df2014.rename(columns={'State': 'State', 'Adult Obesity Rate 2014': 'Obsesity rate-2014'})
Obesity_df2014.head()
```

```python
# merged table for states, datapoints are Obesity rate and mortality,
Corr_df = pd.DataFrame(pd.merge(Obesity_df2014,mortality2014, on = "State"))
Corr_dfcleaned = Corr_df[['Obsesity rate-2014','Mortality']]
Corr_dfcleaned.head()
```

```python
Corr_dfcleaned.corr(method='pearson')
```
